```python
from openpyxl import Workbook
wb = Workbook()
```

```python
#utworzenie trzech arkuszy
gielda = wb.create_sheet("Giełda")
linki = wb.create_sheet("Linki")
filmweb = wb.create_sheet("Filmweb")
```

```python
wb.save('Lab4ex.xlsx')
```

```python
import requests
from bs4 import BeautifulSoup

odpowiedz = requests.get("https://www.filmweb.pl/film/Mroczny+Rycerz+powstaje-2012-506756")
soup = BeautifulSoup(odpowiedz.text, 'html.parser')
ocena = float(soup.find("span", itemprop="ratingValue").text.replace(",", "."))
rezyseria = soup.find(text="reżyseria:").next_element.find("a").text
premiera = soup.find(text="premiera:").next_element.find("a").text
boxoffice = soup.find(text="boxoffice:").next_element.text[1:].replace(" ", "")

print(f"ocena: {ocena}, rezyseria:{rezyseria}, premiera: {premiera}, boxoffice:{boxoffice}")

```

```python
filmweb["A1"] = ocena
filmweb["A2"] = rezyseria
filmweb["A3"] = premiera
filmweb["A4"] = boxoffice
```

```python
wb.save('Lab4ex.xlsx')
```
